```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("UttarPradeshElections2022.csv")
```

```python
# codecs provides access to the internal Python codec registry
import codecs

from deep_translator import GoogleTranslator

# This is to analyse the sentiment of text
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
bjp =0
cong = 0
for sentence in data:   
    translated_text = GoogleTranslator(source='auto', target='en').translate(sentence) 
    analyzer = SentimentIntensityAnalyzer()
    sentiment_dict = analyzer.polarity_scores(translated_text)  
    k = sentence.find("#bjp")
    ks = sentence.find("#BJP")
    if(k >=0 or ks>=0):
        bjp+=sentiment_dict['compound']
        print(f"bjp:  {bjp}")
    c = sentence.find("#bjp")
    cs = sentence.find("#BJP")
    if(c >=0 or cs>=0):
        cong+=sentiment_dict['compound']
        print(f"congress:  {cong}")
        
print(cong)
print(bjp)
```
